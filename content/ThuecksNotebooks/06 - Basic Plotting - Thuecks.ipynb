### Estimating pi through Monte Carlo sampling

Determine the fraction of points randomly chosen between (0,0) and (1,1) falling within the quarter circle:

$$ \Large y = (1-x^2)^{1/2} $$

We previously used this to estimate $\pi$. 

Now, Show your data visually by plotting the quarter arc as a black line, data inside the arc as blue markers, and data outside the arc as red markers.

*Hint: Remember that `x=random.random()` (from the `random` package) assigns a random value to $x$ that lies between 0 and 1.*



```python
import random
import matplotlib.pyplot as plt
import numpy as np

N = 1000
c = 0 # Counter for number of coords inside arc

# Generate empty lists for data to go into
xlist = []
ylist = []
ilist = []

xarc = np.linspace(0,1,1000)
yarc = np.sqrt(1-xarc**2)

# Use for loop to randomly select (x,y) coords and add to list
for i in range(N):
    x = random.random()
    y = random.random()
    # Count number of coords within circle
    if (x**2 + y**2) <= 1:
        c += 1
        ilist.append(1)
    else:
        ilist.append(0)
    xlist.append(x)
    ylist.append(y)
    
xdata = np.array(xlist)
ydata = np.array(ylist)
idata = np.array(ilist)

# Examine ratios to get pi
pi_est = 4*c/N
print(pi_est)

# Plot inside circle as red, outside as black
plt.figure(dpi=100)
plt.plot(xarc,yarc)
plt.xlim([0,1])
plt.ylim([0,1])

plt.scatter(xdata[idata==1],ydata[idata==1],color='red',marker='.')
plt.scatter(xdata[idata==0],ydata[idata==0],color='black',marker='.')
#print(idata==1)
```

### Drawing the semi-empirical mass formula using previous homework problem.
In nuclear physics, the semi-empirical mass formula is a formula
for calculating the approximate nuclear binding energy $B$ of an atomic
nucleus with atomic number $Z$ and mass number $A$:
$$B = a_1 A - a_2 A^{2/3} - a_3 {Z^2\over A^{1/3}}
    - a_4 {(A - 2Z)^2\over A} + {a_5\over A^{1/2}}\,,$$
where, in units of millions of electron volts, the constants are
$a_1=15.8$, $a_2=18.3$, $a_3=0.714$, $a_4=23.2$, and
$$a_5 = \left\lbrace\begin{array}{ll}
      0     &\quad\mbox{if $A$ is odd,} \\
      12.0  &\quad\mbox{if $A$ and $Z$ are both even,} \\
      -12.0 &\quad\mbox{if $A$ is even and $Z$ is odd.}
      \end{array}\right.$$

Consider the homework question completed earlier. Plot the maximal value of $B/A$ for each value of Z.

The homework question solution is provided. How would you accomplish the plotting task above?

```python
def bindenergy(A,Z):
    a1 = 15.8
    a2 = 18.3
    a3 = 0.714
    a4 = 23.2
    if A%2==1:
        a5 = 0.0
    elif A%2==0 and Z%2==0:
        a5 = 12.0
    else:
        a5 = -12.0
    
    return (a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(A-2*Z)**2/A + a5/A**(0.5))

########### Main program ######################

Zmax = 100
for Z in range(1,Zmax+1):
    maxbpn = 0.0
    maxA = 0
    
    for A in range(Z,3*Z+1):
        bpn = bindenergy(A,Z)/A
        if bpn > maxbpn:
            maxbpn = bpn
            maxA = A
    print("For Z = {0}, the most stable nucleus is when A = {1} where binding energy per nucleon is {2} MeV".format(Z,maxA,maxbpn))
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>

#### Solution:
I've created numpy arrays to hold $Z$ values, max$(B/A)$ for each $Z$, and the value of $A$ at which we see max$(B/A)$ for each Z. Note that array indices start at 0 while the minimum value of $Z$ is 1, hence the `Z-1` indices in the `for` loop.

```python
def bindenergy(A,Z):
    a1 = 15.8
    a2 = 18.3
    a3 = 0.714
    a4 = 23.2
    if A%2==1:
        a5 = 0.0
    elif A%2==0 and Z%2==0:
        a5 = 12.0
    else:
        a5 = -12.0
    
    return (a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(A-2*Z)**2/A + a5/A**(0.5))

########### Main program ######################
import numpy as np
import matplotlib.pyplot as plt

Zmax = 100

Z_arr = np.zeros(Zmax)
A_arr = np.zeros(Zmax)
bepn_arr = np.zeros(Zmax)

for Z in range(1,Zmax+1):
    maxbpn = 0.0
    maxA = 0
    
    for A in range(Z,3*Z+1):
        bpn = bindenergy(A,Z)/A
        if bpn > bepn_arr[Z-1]:
            bepn_arr[Z-1] = bpn
            A_arr[Z-1] = A
    Z_arr[Z-1] = Z

###### Plot the results ###########
fig = plt.figure(dpi=150)

plt.plot(Z_arr,bepn_arr,'m-')
plt.scatter(Z_arr,bepn_arr, c='blue',s=7)
plt.xlabel('Z')
plt.ylabel('max(B/A) (MeV)')
plt.title('Semi-Empirical Mass Binding Formula')

plt.show()
```

## Heatmaps

There may be many instances when you would like to visualize the behavior of a quantity as a function of multiple variables (e.g. x,y or x,t). In situations like this, there are methods to plot arrays of data. One such method is called a heatmap (or density plot).

The textbook uses a function from matplotlib.pyplot called `imshow`. I actually prefer `pcolormesh` (more flexible than `imshow` and faster than `pcolor`).

```python
import numpy as np
import matplotlib.pyplot as plt

#Load data from Thuecks 2017 Physics of Plasmas article (fig 10)
f = np.loadtxt('datasets/2017_fig10_freq.csv',delimiter=',')
kperp = np.loadtxt('datasets/2017_fig10_kperp.csv',delimiter=',')
data = np.loadtxt('datasets/2017_fig10_normcoh.csv',delimiter=',')

#Plot a heatmap 
plt.figure(1,dpi=100)
plt.pcolormesh(kperp,f,data.transpose()) # Flip data so that kperp is along x-axis ..... 
plt.xlabel('$k_\perp$ (cm$^{-1}$)')
plt.ylabel('Frequency (Hz)')
plt.title('Normalized Coherence')
```
